```python
#import libraries
import pandas as pd
import zipfile
```

# Read zip files and convert them into Dataframe
---
- pick the trip data in December 2022 and January 2023
- decided to analysis the Christmas and New Year's tripdata to see the citibike usage in the public holidays
- where could be improve for 2023 Christmas and New Year's citibike operation

```python
#file path
dec_bike = "Data/JC-202212-citibike-tripdata.zip"
```

```python
# open zipped dataset
with zipfile.ZipFile(dec_bike) as z:
   # open the csv file in the dataset
   with z.open("JC-202212-citibike-tripdata.csv") as f:
       
      # read the dataset
    dec_tripdata = pd.read_csv(f)
       
      # display dataset
    print(dec_tripdata)
```

```python
#file path
jan_bike = "Data/JC-202301-citibike-tripdata.zip"
```

```python
# open zipped dataset
with zipfile.ZipFile(jan_bike) as z:
   # open the csv file in the dataset
   with z.open("JC-202301-citibike-tripdata.csv") as f:
       
      # read the dataset
    jan_tripdata = pd.read_csv(f)
       
      # display dataset
    print(jan_tripdata)
```

```python
#convert to DataFrame
dec_df=pd.DataFrame(dec_tripdata)
dec_df
```

```python
#Convert to dataframe
jan_df = pd.DataFrame(jan_tripdata)
jan_df
```

### narrow down the time range in each dataframe

```python
from datetime import datetime
```

```python
#data types in dec_df
dec_df.dtypes
```

```python
#change the type of time
dec_df["started_at"] = pd.to_datetime(dec_df["started_at"])
dec_df["ended_at"] = pd.to_datetime(dec_df["ended_at"])
```

```python
#make sure it changed
dec_df.dtypes
```

```python
#data types in jan_df
jan_df.dtypes
```

```python
#change the type of time
jan_df["started_at"] = pd.to_datetime(jan_df["started_at"])
jan_df["ended_at"] = pd.to_datetime(jan_df["ended_at"])
```

```python
#make sure it changed
jan_df.dtypes
```

```python
#Christmas time data
chris_data = dec_df.loc[dec_df['started_at'].between('2022-12-23','2022-12-26')]
chris_data
```

```python
#new year data
new_year_eve_data = dec_df.loc[dec_df['started_at'].between('2022-12-30','2022-12-31')]
new_year_eve_data
```

```python
new_year_data = jan_df.loc[jan_df['started_at'].between('2023-01-01','2023-01-02')]
new_year_data
```

```python
#merge new year eve and new year data
merge_new_year = pd.concat([new_year_eve_data, new_year_data])
merge_new_year
```

```python
#output into .csv file
merge_new_year.to_csv('merge_new_year.csv')
chris_data.to_csv('chris_data.csv')
```
